```python
# Create an in-memory SQLite database and write the DataFrame to a table
conn = sqlite3.connect(":memory:")
df.to_sql("sales", conn, index=False, if_exists="replace")
```

```python
# Run the updated SQL query
query = """
SELECT
  [Product Category] AS category,
  SUM(Quantity) AS total_quantity,
  SUM([Total Amount]) AS total_sales
FROM sales
GROUP BY [Product Category]
ORDER BY total_sales DESC
"""
summary_df = pd.read_sql_query(query, conn)
print(summary_df)
```

```python
# STEP 6: Bar chart of Total Sales by Product Category
summary_df.plot(kind='bar', x='category', y='total_sales', legend=False, color='steelblue')
plt.title("Total Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# SQL query to calculate profit and profit margin by category
query = """
SELECT
    [Product Category] AS Category,
    SUM(Quantity * [Price per Unit]) AS Total_Sales,
    SUM(Quantity * ([Price per Unit] / 1.3)) AS Estimated_Cost,
    SUM(Quantity * [Price per Unit]) - SUM(Quantity * ([Price per Unit] / 1.3)) AS Profit,
    ROUND(
        (SUM(Quantity * [Price per Unit]) - SUM(Quantity * ([Price per Unit] / 1.3))) * 100.0
        / NULLIF(SUM(Quantity * [Price per Unit]), 0), 2
    ) AS Profit_Margin_Percent
FROM sales
GROUP BY [Product Category]
ORDER BY Profit_Margin_Percent DESC
"""

# Execute query and store result in DataFrame
profit_margin_df = pd.read_sql_query(query, conn)
profit_margin_df
```

```python
# @title Profit_Margin_Percent vs Total_Sales

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Profit_Margin_Percent']
  ys = series['Total_Sales']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = profit_margin_df.sort_values('Profit_Margin_Percent', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Category')):
  _plot_series(series, series_name, i)
  fig.legend(title='Category', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Profit_Margin_Percent')
_ = plt.ylabel('Total_Sales')
```
